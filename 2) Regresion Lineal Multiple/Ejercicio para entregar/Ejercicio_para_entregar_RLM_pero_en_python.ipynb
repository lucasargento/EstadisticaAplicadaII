```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm
import pandas as pd
```

```python
df = pd.read_excel("Datos guia RLM 1.xlsx", sheet_name="Problema7")
```

```python
df.head()
```

```python
import plotly.graph_objs as go

peso = df["Peso del pescado en gramos"]
ancho = df["Ancho en pulgadas"]
longitud = df["Longitud vertical en pulgadas"]

# Fit a linear regression model

# Las x son mis features o variables explicativas. notar que es una matriz, RLM
X = sm.add_constant(np.column_stack((ancho, longitud)))
# creo el modelo de Y(X)
modelo_rlm = sm.OLS(peso, X).fit()
coeficientes_rlm = modelo_rlm.params

# Ploteo los datos y el plano de regresion.

# Define points on the plane (using the range of your data)
ancho_grid, longitud_grid = np.meshgrid(np.linspace(min(ancho), max(ancho), 10),
                                        np.linspace(min(longitud), max(longitud), 10))
z_plane = coeficientes_rlm[0] + coeficientes_rlm[1] * ancho_grid + coeficientes_rlm[2] * longitud_grid

# Create scatter plot trace
scatter_trace = go.Scatter3d(x=ancho, y=longitud, z=peso, mode='markers', marker=dict(color='blue', size=5),
                             name='Data')

# Create surface plot trace for the regression plane
surface_trace = go.Surface(x=ancho_grid, y=longitud_grid, z=z_plane, opacity=0.5, colorscale='Reds',
                           name='Regression Plane')

# Create layout
layout = go.Layout(title='Peso vs Ancho y Longitud',
                   scene=dict(xaxis=dict(title='Ancho'),
                              yaxis=dict(title='Longitud'),
                              zaxis=dict(title='Peso')))

# Create figure
fig = go.Figure(data=[scatter_trace, surface_trace], layout=layout)

# Show interactive plot
fig.show()
```
